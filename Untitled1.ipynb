```
from matplotlib.pyplot import cm 
from mpl_toolkits.mplot3d import Axes3D
import scipy.linalg as sla
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import numpy as np
def pot_eng(par):
    n=int(0.5*len(par))
    u=0.0
    for i in range(n-1):
        for j in range(i+1,n):
            thi,thj=par[i],par[j]
            phi,phj=par[i+n],par[j+n]
            r=rij(thi,thj,phi,phj)
            u+=1/r
    return u

def rij(thi,thj,phi,phj):
    a= (( np.sin(thi)*np.cos(phi) ) -  ( np.sin(thj)*np.cos(phj) ))
    b=((np.sin(thi)*np.sin(phi)) - (np.sin(thj)*np.sin(phj)))
    c=((np.cos(thi)-np.cos(thj))) 
    return (a**2+b**2+c**2)**0.5
n=10
par=[0.1*i for i in range(2*n)]
rs=minimize(pot_eng,par,method='powell',options={'xtol':1e-5,'disp':True})
par=rs.x
for i in range(n-1):
    for j in range(i+1,n):
        thi,thj=par[i],par[j]
        phi,phj=par[i+n],par[j+n]
        r=rij(thi,thj,phi,phj)
        print('r%d%d = %.5f'%(i,j,r))
xx,yy,zz=[],[],[]
for i in range(n):
    th,ph=par[i],par[i+n]
    xx.append(np.sin(th)*np.cos(ph))
    yy.append(np.sin(th)*np.sin(ph))
    zz.append(np.cos(th))
th,ph=np.meshgrid(np.linspace(0,np.pi,100),np.linspace(0,2*np.pi,100))
x=np.sin(th)*np.cos(ph)
y=np.sin(ph)*np.sin(th)
z=np.cos(th)

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot_surface(x,y,z,rstride=1,cstride=1,color='c',alpha=0.3,linewidth=0.0)
ax.scatter(xx,yy,zz,s=20)
ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
plt.title('The %d particle system'%n)
plt.show()

        

        
        
        
```
